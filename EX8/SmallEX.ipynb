## 01 Exercise

* Use the `CountVectorizer` from `sklearn.feature_extraction` to read the book `data/moby_dick.txt`
  * How many times does the word 'wood' appear?

````python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np

text = open("../../data/moby_dick.txt").read().replace("\n"," ")
vectorizer = CountVectorizer()

matrix = vectorizer.fit_transform([text])
matrix.toarray()
counts_df = pd.DataFrame(matrix.toarray(), columns = vectorizer.get_feature_names())
counts_df.T.sort_values(by=0, ascending=False).head(8)
counts_df['wood']
````
