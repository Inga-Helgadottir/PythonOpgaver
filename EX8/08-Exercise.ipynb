```python
import requests
import pandas as pd
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
%matplotlib inline

data = 'https://think.cs.vt.edu/corgis/datasets/csv/cars/cars.csv'
dataCSV = pd.read_csv(data)
df = pd.DataFrame(dataCSV)
print(df) 
# print(df.columns) # find out column names
```

# Feature engineering

Reduce the dataset by:
    Make == Honda
    Fueltype == Gasoline
    Remove outliers. Hint: df[column name'] < df['column name'].quantile(0.90)
    Only use 2 features: 'Fuel Information.Highway mpg' and 'Engine Information.Engine Statistics.Horsepower'
    Normalize data (reduce values to be between 0 and 1)

```python
df = df[(df['Identification.Make']=="Honda") & (df['Fuel Information.Fuel Type']=="Gasoline")]
print(df) 
```

```python
df = df[(df['Fuel Information.Highway mpg'] < df['Engine Information.Engine Statistics.Horsepower'].quantile(0.90))]
print(df) 
```

```python
df = df[['Fuel Information.Highway mpg', 'Fuel Information.Fuel Type']]
print(df)
```
