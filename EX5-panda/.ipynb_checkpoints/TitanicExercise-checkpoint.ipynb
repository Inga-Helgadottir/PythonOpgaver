    6. Get only last name by the use of str method split() on comma. https://pandas.pydata.org/docs/user_guide/text.html#splitting-and-replacing-strings
    7. Get only those passengers that had one of these ticket numbers: [350406,248706,382652,244373,345763,2649,239865]

```python
#1. read titanic data from the data/titanic_train.csv file into a pandas dataframe
import pandas as pd
filename = "../../data/titanic_train.csv"
df = pd.read_csv(filename)
# print(df)

#2. set index of the dataframe to passenger id
df2 = df.set_index('PassengerId')
#df2

#3. sort the dataframe by 'fare' (price of ticket)
df2_sorted = df2.sort_values('Fare',ascending=True)
#df2_sorted

#4. write a function to reformat a name to only contain "firstname lastname" (look for current format in the name column)
def reformat_name(name):
    first_name = name.split(', ')[1].split(' ')[1].strip()
    last_name = name.split(',')[0].strip()
    
    return f'{first_name} {last_name}'

#5. use df.apply() to Change all names to the "firstname lastname" format
df['Name'] = df['Name'].apply(reformat_name)
#df

#6. Get only last name by the use of str method split() on comma.
def reformat_name_last(name):
    last_name = name.split(' ')
    return last_name[1]

df['Name'] = df['Name'].apply(reformat_name_last)
#df

#7. Get only those passengers that had one of these ticket numbers: [350406,248706,382652,244373,345763,2649,239865]
ticketNumbers = ['350406','248706','382652','244373','345763','2649','239865']
#df2.loc[df2['Ticket'].isin(ticketNumbers) == False]
#df
getPassWithTicketNbr = df[df['Ticket'].isin(ticketNumbers)]
# print(getPassWithTicketNbr)

#8. Remove all rows without an age
removeNaNFromAge_df = df[df['Age'].notnull()]
# print(filteredByAge_df)

#9. Convert age values to integer
# convertingToInt = df[['Age']].apply(np.int64)
convertingToInt = removeNaNFromAge_df[['Age']].astype(int)
# print(convertingToInt)

#10. get only the children from the dataset
onlyChildren = removeNaNFromAge_df[removeNaNFromAge_df['Age']<19]
# print(onlyChildren)

#11. Filter to get only passengers with a valid cabin
validCabinNbr = df[df['Cabin'].notnull()]
# print(validCabinNbr)

#12. Change Embarked to full name of port: C = Cherbourg; Q = Queenstown; S = Southampton
changeEmbarked = df.replace({"C":"Cherbourg", "Q":"Queenstown", "S":"Southampton"})
# print(changeEmbarked)

#13. create new column: 'relations' that adds sibling/spouse value with parent/child values
df['Relations'] = df['SibSp']/df['Pclass']
#df

#14. Is there a relationship between how many relations and chances of survival?
corr = df['Relations'].corr(df['Survived'])
# print("Correlation between Relations and Survived is: ", round(corr, 2))
#corr
```
